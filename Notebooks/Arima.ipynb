```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
```

```python
# Load the data from CSV
data1 = pd.read_csv('/content/morocco 2021-01-01 to 2021-12-31.csv', usecols=['datetime', 'temp'])
data2 = pd.read_csv('/content/morocco 2022-01-01 to 2022-12-31.csv', usecols=['datetime', 'temp'])
data3 = pd.read_csv('/content/morocco 2023-01-01 to 2023-05-01.csv', usecols=['datetime', 'temp'])

data = pd.concat([data1, data2, data3], axis=0)
data.reset_index(drop=True, inplace=True)
data
```

```python
# Convert datetime column to datetime object
data['datetime'] = pd.to_datetime(data['datetime'])
data.set_index('datetime', inplace=True)
```

```python
len(data)*0.8
```

```python
# Resample the data to daily frequency and fill missing values (if any)
data = data.resample('D').mean().fillna(method='ffill')
```

```python
# Split data into train and test sets
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]
```

```python
# Fit ARIMA model
# NO multivariate time series forecasting

model = ARIMA(train['temp'], order=(150,1,0))
model_fit = model.fit()

forecast = model_fit.forecast(steps=len(test))
forecast_index = pd.date_range(start=test.index[0], periods=len(test), freq='D')
forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=['temp'])
```

```python
# Plotting
plt.figure(figsize=(10,6))
plt.plot(train.index, train['temp'], label='Training Data')
plt.plot(test.index, test['temp'], label='Test Data')
plt.plot(test.index, forecast, label='ARIMA Forecast')
plt.title('ARIMA Weather Forecasting')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```
